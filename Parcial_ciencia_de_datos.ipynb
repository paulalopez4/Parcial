```
import pandas as pd
import numpy as np

# Dataset base
data = {
    'Producto': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Ventas': [120, 340, 290, 410, 150, 380],
    'Costo': [80, 200, 180, 260, 90, 240],
    'Categoria': ['Tecnología', 'Hogar', 'Tecnología', 'Hogar', 'Oficina', 'Tecnología'],
    'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo']
}
df = pd.DataFrame(data)

print("=" * 60)
print("DATASET BASE")
print("=" * 60)
print(df)
print("\n")


# PREGUNTA 1 – NumPy (Vectores y operaciones)

print("=" * 60)
print("PREGUNTA 1 - NumPy (Vectores y operaciones)")
print("=" * 60)

# Crear arreglo de NumPy con las ventas
ventas_array = np.array(df['Ventas'])
print(f"Arreglo de ventas: {ventas_array}")

# a) Calcular promedio, máximo y mínimo
promedio = np.mean(ventas_array)
maximo = np.max(ventas_array)
minimo = np.min(ventas_array)

print(f"\nPromedio de ventas: {promedio}")
print(f"Valor máximo: {maximo}")
print(f"Valor mínimo: {minimo}")

print("\n EXPLICACIÓN:")
print("- np.mean(): Calcula el promedio aritmético sumando todos los valores y dividiéndolos entre el total de elementos")
print("- np.max(): Retorna el valor más alto del arreglo")
print("- np.min(): Retorna el valor más bajo del arreglo")
print("\n")
```

```
# PREGUNTA 2 – NumPy (Condiciones)

print("=" * 60)
print("PREGUNTA 2 - NumPy (Condiciones)")
print("=" * 60)

# Crear arreglo con 1 si venta > 300, 0 en caso contrario (sin usar for)
ventas_mayor_300 = np.where(ventas_array > 300, 1, 0)

print(f"Ventas: {ventas_array}")
print(f"Ventas > 300: {ventas_mayor_300}")
print("\n EXPLICACIÓN:")
print("Se usa np.where() que evalúa la condición para cada elemento del arreglo")
print("de forma vectorizada (sin ciclos). Retorna 1 si la condición es True, 0 si es False")
print("\n")
```

```
# PREGUNTA 3 – Pandas (Columnas calculadas)

print("=" * 60)
print("PREGUNTA 3 - Pandas (Columnas calculadas)")
print("=" * 60)

# Agregar columna Ganancia
df['Ganancia'] = df['Ventas'] - df['Costo']

print("DataFrame con columna Ganancia:")
print(df.head())

print("\nProductos con ganancia mayor a 100:")
productos_ganancia_alta = df[df['Ganancia'] > 100]
print(productos_ganancia_alta[['Producto', 'Ganancia']])
print("\n")
```

```
# PREGUNTA 4 – Pandas (Filtros y condiciones)

print("=" * 60)
print("PREGUNTA 4 - Pandas (Filtros y condiciones)")
print("=" * 60)

# Filtrar: Categoría Tecnología Y Ventas > 300
tecnologia_ventas_altas = df[(df['Categoria'] == 'Tecnología') & (df['Ventas'] > 300)]

print("Productos de Tecnología con ventas > 300:")
print(tecnologia_ventas_altas[['Producto', 'Ventas', 'Ganancia']])
print("\n")
```

```
# PREGUNTA 5 – Pandas (Agrupaciones)

print("=" * 60)
print("PREGUNTA 5 - Pandas (Agrupaciones)")
print("=" * 60)

# Agrupar por Mes
agrupado_mes = df.groupby('Mes').agg({
    'Ventas': 'sum',
    'Ganancia': 'mean'
}).round(2)

print("Ventas totales y ganancia promedio por mes:")
print(agrupado_mes)

print("\n EXPLICACIÓN - Diferencia entre sum() y mean():")
print("- sum(): Suma todos los valores del grupo. Útil para obtener totales acumulados")
print("- mean(): Calcula el promedio aritmético del grupo. Útil para obtener valores representativos")
print("\nEn este caso:")
print("- Ventas totales (sum): Muestra cuánto se vendió en total cada mes")
print("- Ganancia promedio (mean): Muestra la ganancia típica por producto en cada mes")
print("\n")

```

```
# PREGUNTA 6 - Gráfico de barras de ventas por producto
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.DataFrame(data)

# Add the 'Ganancia' column, as it's used in previous cells and might be needed for consistency
df['Ganancia'] = df['Ventas'] - df['Costo']

plt.figure(figsize=(7, 5))
barras = plt.bar(df['Producto'], df['Ventas'], color='pink')

# Título y etiquetas de ejes
plt.title("Ventas por Producto")
plt.xlabel("Producto")
plt.ylabel("Ventas")

plt.xticks(rotation=45)

# Etiquetas de valor
for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        altura,
        f'{altura}',
        ha='center',
        va='bottom'
    )

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```
# PREGUNTA 7 - Gráfico comparativo de Ventas vs COsto
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame(data)

plt.figure(figsize=(7, 5))
plt.plot(df['Producto'], df['Ventas'],
         color='purple', marker='o', linewidth=2, label='Ventas')

plt.plot(df['Producto'], df['Costo'],
         color='blue', marker='s', linewidth=2, label='Costo')

# Títulos y etiquetas
plt.title('Comparación de Ventas vs Costo por Producto')
plt.xlabel('Producto')
plt.ylabel('Monto')
plt.legend()


plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

print("\n EXPLICACIÓN - Tendencia e interpretación del gráfico:")
print("En todos los productos (A–F), la línea de Ventas se mantiene por encima de la línea de Costo,")
print ("lo que indica que todos los productos son rentables.")
print("El producto D presenta el valor más alto tanto en ventas como en costos, así como el mayor")
print("margen de ganancia y rentabilidad. Por el contrario, los productos A y E tienen los valores")
print ("más bajos de ventas y costo, así como la diferencia entre estos es cercana, lo que sugiere")
print ("que la ganancia con su venta es baja.")
print("\n")
```

```
# PREGUNTA 8. Gráfico de cajas y bigotes de las ventas por categoría
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame(data)

sns.set(style="whitegrid")

palette = {'Tecnología': 'skyblue', 'Hogar': 'yellow', 'Oficina': 'orange'}

plt.figure(figsize=(7,5))
sns.boxplot(
    x='Categoria',
    y='Ventas',
    hue='Categoria',
    data=df,
    palette=palette,
    dodge=False)

plt.legend([],[], frameon=False)

plt.title('Distribución de Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas')

plt.tight_layout()
plt.show()


print("\n EXPLICACIÓN - Tendencia e interpretación del gráfico:")
print("La caja de Tecnología es mucho más alta y los bigotes son considerablemente más extensos")
print("en relación con las de hogar y oficina. Por lo tanto, el gráfico sugiere que la categoría")
print ("de Tecnología es la que presenta una mayor variabilidad y dispersión en sus datos de ventas.")
print ("Mientras que en la de Hogar las ventas son más parecidas o similares entre los productos")
print ("de esa categoría. Por otro lado, Oficina solo presenta un dato entonces no habrá dispersión ")
print("\n")

```

```
#PUNTO 9 - Scatterplot de Ventas y Ganancias por categoría
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#
df = pd.DataFrame(data)

# Calculo de la ganancia
df['Ganancia'] = df['Ventas'] - df['Costo']

sns.set(style="whitegrid")

# Scatterplot: Ventas vs Ganancia por Categoría
plt.figure(figsize=(7, 5))
sns.scatterplot(
    data=df,
    x='Ventas',
    y='Ganancia',
    hue='Categoria',
    style='Categoria',
    s=100,
    palette='Set2')

plt.title('Relación entre Ventas y Ganancia por Categoría')
plt.xlabel('Ventas')
plt.ylabel('Ganancia')

plt.tight_layout()
plt.show()


print("\n EXPLICACIÓN - Tendencia e interpretación del gráfico:")
print("Se observa que la relación observada entre ventas y ganancia es directamente proporcional")
print("dado que, cuando las ventas aumentan, las ganancias también lo hacen. De esta manera,")
print ("la gráfica muestra un comportamiento ascendente que significa un margen de ganancia positivo")
print ("y que al aumentar el volumen de ventas se genera un mayor margen de rentabilidad")
print ("La categoría de Hogar es la que más ganancias generan, mientras que la de Oficina")
print ("por sus bajas ventas es la que ha generado menos beneficios económicos")
print("\n")
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.DataFrame(data)

# Calcular Ganancia
df['Ganancia'] = df['Ventas'] - df['Costo']

#Gráfico de barras: Ganancia por Producto

palette_producto = {'A':'skyblue','B':'lightgreen','C':'salmon','D':'purple','E':'orange','F':'pink'}

plt.figure(figsize=(8,5))
sns.barplot(
    x='Producto',
    y='Ganancia',
    hue='Producto',
    data=df,
    palette=palette_producto,
    dodge=False,
    legend=False)
plt.title('Ganancia por Producto')
plt.xlabel('Producto')
plt.ylabel('Ganancia')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Gráfico de barras: Ganancia total por Categoría
ganancia_por_categoria = df.groupby('Categoria')['Ganancia'].sum().reset_index()
palette_categoria = {'Tecnología': 'skyblue', 'Hogar': 'lightgreen', 'Oficina': 'salmon'}

plt.figure(figsize=(6,5))
sns.barplot(
    x='Categoria',
    y='Ganancia',
    hue='Categoria',
    data=ganancia_por_categoria,
    palette=palette_categoria,
    dodge=False,
    legend=False)
plt.title('Ganancia Total por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ganancia Total')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Pandas para decisión comercial
total_ganancia = df['Ganancia'].sum()
df['%_contrib_ganancia'] = (df['Ganancia'] / total_ganancia * 100).round(2)

df_resumen = df[['Producto', 'Categoria', 'Ventas', 'Costo', 'Ganancia', '%_contrib_ganancia']].sort_values(by='Ganancia', ascending=False)

print("Resumen de productos para decisión comercial:")
print(df_resumen)


print("\n EXPLICACIÓN - Análisis de resultados:")
print("a) El producto más rentable es el D, ya que presenta la mayor ganancia individual")
print("de todo el portafolio, lo que lo convierte en un producto estratégico para priorizar.")
print("b) Las categorías que generan mayor ganancia total son Hogar y Tecnología,")
print("ambas concentrando la mayor parte de los beneficios económicos. Por su parte, Oficina")
print("aporta poco a la rentabilidad total debido a sus bajas ventas.")
print("c) La decisión comercial recomendada es enfocar la inversión y promoción en los")
print("productos más rentables como D, B y F, así como en las categorías que aportan mayor ganancia,")
print("mientras se evalúa optimizar o ajustar la categoría Oficina para mejorar su desempeño.\n")



```
